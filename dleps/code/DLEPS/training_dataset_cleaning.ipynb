## Select the corresponding **CatalogNumber** for NA smiles to retrieve

```python
# Select the CatalogNumber
num_catalog_list = []
for i in range(len(df_fda_selected)):
    if df_fda_selected.SMILES[i] is np.nan:
        num_catalog_list.append(df_fda_selected.CatalogNumber[i])
print(len(num_catalog_list))
print(num_catalog_list)
```

```python
# Retrieve the missing SMILES from url given
import requests
from bs4 import BeautifulSoup

smiles_list = []
badi = []

for i in range(len(num_catalog_list)):
    url = 'http://www.apexbt.com/search.php?catalog='
    url = url + str(num_catalog_list[i])
    # print(url)
    try: 
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        row = soup.find('td', string='Canonical SMILES').find_next_sibling('td')
        smiles_list.append(row.text)
    except:
        print("Unexpected error occured at " + str(i) + "th iteration, catalog number: " + str(num_catalog_list[i]) + ".")
        badi.append(i)
        smiles_list.append('NA')

print(len(smiles_list))
```

```python
print(len(smiles_list), len(badi))
```

```python
# Create dataframe with catalog number and smiles
df_smiles = pd.DataFrame({'CatalogNumber': num_catalog_list, 'SMILES': smiles_list})
df_smiles
```

```python
# Add the smiles to the original dataframe
for i in range(len(df_fda_selected)):
    for j in range(len(df_smiles)):
        if df_fda_selected.CatalogNumber[i] == df_smiles.CatalogNumber[j]:
            df_fda_selected.SMILES[i] = df_smiles.SMILES[j]
df_fda_selected
```

```python
# Count the NA counts in smiles
na_smiles_count = sum(df_fda_selected.SMILES == 'NA')
na_smiles_count
```
