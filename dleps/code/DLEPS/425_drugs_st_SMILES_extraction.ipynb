```python
import pandas as pd

df1 = pd.read_csv('../../data/gen_exp_drugs.csv')
df1
```

```python
df2 = pd.read_csv('../../data/gene_expression_data/train_chembl_merged_smiles.csv')
df2 = df2.drop_duplicates(subset='standardized_SMILES').drop_duplicates(subset='pref_name').reset_index(drop=True)
df2
```

```python
# Extract standardized SMILES for 425 drugs
standardized_SMILES = []
for i in range(len(df1)):
    for j in range(len(df2)):
        if df1['drug_name'][i].lower() == df2['pref_name'][j].lower():
            standardized_SMILES.append(df2['standardized_SMILES'][j])
print(len(standardized_SMILES))
standardized_SMILES
```

```python
df1['standardized_SMILES'] = standardized_SMILES
df1.to_csv('../../data/gen_exp_drugs_st_SMILES.csv', index=False)
df1
```
