```python
# Import libraries

import numpy as np
import pandas as pd
import h5py
```

```python
# Load the gene exp data

h5f = h5py.File('../../data/gene_exp_data.h5', 'r')
data = h5f['data'][:]

data.shape
```

```python
# Load the unique drugs in gene expression data
unique_drugs = pd.read_csv('../../data/unique_drugs.csv')
unique_drugs
```

```python
# Load the training dataset
df_train = pd.read_csv('../../data/train_data_with_mean_values.csv')
df_train
```

```python
# Load train chembl merged dataset

df_merged = pd.read_csv('../../data/gene_expression_data/train_chembl_merged_smiles.csv')
df_merged
```

```python
# Add mean values to merged dataset

labels = {}
for drug_1 in df_merged.DRUG_NAME:
    for j in range(len(df_train)):
        if drug_1.lower() == df_train['DRUG NAME'][j].lower():
            labels[df_train['DRUG NAME'][j]] = df_train['MDA7+D1+Mean'][j]
df_labels = pd.DataFrame(list(labels.items()), columns=['Drug Name', 'MDA7+D1+Mean'])
df_labels
```

```python
df_merged_labels = df_merged.merge(df_labels, left_on="DRUG_NAME", right_on="Drug Name")
df_merged_labels
```

```python
scores = np.zeros(len(unique_drugs))
for i in range(len(unique_drugs)):
    for j in range(len(df_merged_labels)):
        if df_merged_labels['pref_name'][j].lower() == unique_drugs['drug_name'][i]:
            scores[i] = df_merged_labels['MDA7+D1+Mean'][j] 
len(scores)
```

```python
# Split train, test
perm = np.arange(len(scores))
np.random.shuffle(perm)
data = data[perm]
labels = scores[perm]
TEST_SIZE = 100
data_train = data[TEST_SIZE:]
labels_train = labels[TEST_SIZE:]
data_test = data[:TEST_SIZE]
labels_test = labels[:TEST_SIZE]
```

```python
import h5py

h5f = h5py.File('../../data/gene_exp_data_train.h5', 'w')
h5f.create_dataset('data', data=data_train)
h5f.close()

h5f = h5py.File('../../data/gene_exp_labels_train.h5', 'w')
h5f.create_dataset('data', data=labels_train)
h5f.close()

h5f = h5py.File('../../data/gene_exp_data_test.h5', 'w')
h5f.create_dataset('data', data=data_test)
h5f.close()

h5f = h5py.File('../../data/gene_exp_labels_test.h5', 'w')
h5f.create_dataset('data', data=labels_test)
h5f.close()
```
