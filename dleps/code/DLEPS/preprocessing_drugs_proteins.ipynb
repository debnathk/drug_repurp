```python
# Load libraries
import pandas as pd
import numpy as np
import h5py
import utils
```

```python
# Load the dataset
df = pd.read_csv('../../data/BindingDB_IC50_updated.csv')
print(df.shape)
df.head()
```

```python
# Segregate drugs, proteins and labels
df_drugs = df['Ligand SMILES']
df_proteins = df['BindingDB Target Chain Sequence']
df_labels = df['IC50 (nM)']
```

```python
# Preprocess drugs
df_drugs
```

```python
# One-hot encoding of drug SMILES

S = pd.Series(df_drugs.unique()).apply(utils.smiles2onehot)
S_dict = dict(zip(df_drugs.unique(), S))
df_drugs = [S_dict[i] for i in df_drugs]
```

```python
one_hot_drugs = np.array(df_drugs)
one_hot_drugs.shape
```

```python
one_hot_drugs
```

```python
# Preprocess proteins
df_proteins
```

```python
# Tokenize amino acid sequences

AA = pd.Series(df_proteins.unique()).apply(utils.protein2onehot)
AA_dict = dict(zip(df_proteins.unique(), AA))
df_proteins = [AA_dict[i] for i in df_proteins]
```

```python
one_hot_proteins = np.array(df_proteins)
one_hot_proteins.shape
```

```python
one_hot_proteins
```

```python
# Preprocess labels
df_labels
```

```python
# Transform labels

L_dict = pd.Series(df_labels)
l_converted = L_dict.apply(lambda l: utils.convert_y_unit(l, from_='nM', to_='p')).tolist()
np.shape(l_converted)
```

```python
l_converted
```

```python
# Save drug encodings

h5f = h5py.File('../../data/drugs.h5', 'w')
h5f.create_dataset('data', data=one_hot_drugs)
h5f.close()
```

```python
h5f = h5py.File('../../data/drugs.h5', 'r')
data = h5f['data'][:]
h5f.close()

data
```

```python
# Save protein encodings

h5f = h5py.File('../../data/proteins.h5', 'w')
h5f.create_dataset('data', data=one_hot_proteins)
h5f.close()
```

```python
h5f = h5py.File('../../data/proteins.h5', 'r')
data = h5f['data'][:]
h5f.close()

data
```

```python
# Save labels

h5f = h5py.File('../../data/pIC50.h5', 'w')
h5f.create_dataset('data', data=l_converted)
h5f.close()
```

```python
h5f = h5py.File('../../data/pIC50.h5', 'r')
data = h5f['data'][:]
h5f.close()

data
```
