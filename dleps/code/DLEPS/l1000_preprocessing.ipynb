```python
gene_expression_df_10um_conc_up = gene_expression_df_10um_conc[gene_expression_df_10um_conc[0].str.contains('up')]
gene_expression_df_10um_conc_up.to_csv('../../data/l1000_cp_10uM_up_filtered.csv', index=False)
gene_expression_df_10um_conc_up
```

```python
gene_expression_df_10um_conc_down = gene_expression_df_10um_conc[gene_expression_df_10um_conc[0].str.contains('down')]
gene_expression_df_10um_conc_down.to_csv('../../data/l1000_cp_10uM_down_filtered.csv', index=False)
gene_expression_df_10um_conc_down
```

```python
# Clean the drug names in the replicates - up

gene_expression_df_10um_conc_up.loc[:, 0] = gene_expression_df_10um_conc_up[0].apply(extract_drug_name)
gene_expression_df_10um_conc_up.to_csv('../../data/l1000_cp_10uM_up_filtered_clean.csv', index=False)
gene_expression_df_10um_conc_up
```

```python
len(set(gene_expression_df_10um_conc_up[0]))
```

```python
# Clean the drug names in the replicates - down

gene_expression_df_10um_conc_down.loc[:, 0] = gene_expression_df_10um_conc_down[0].apply(extract_drug_name)
gene_expression_df_10um_conc_down.to_csv('../../data/l1000_cp_10uM_down_filtered_clean.csv', index=False)
gene_expression_df_10um_conc_down
```

```python
drug_names_unique = np.unique(gene_expression_df_10um_conc_down[0])
drug_names_unique
```

```python
# Create dataset

# Read the landmark_genes CSV file
landmark_genes = pd.read_csv("../../data/landmark_genes.csv", header=None)

# Prepare a list to store the results
data_reg_list = []

# Precompute the count of occurrences for each drug in the down-regulated dataset
drug_counts = gene_expression_df_10um_conc_down[0].value_counts()

# Iterate over each unique drug name
for drug in drug_names_unique:
    # Initialize the DataFrame for this drug with zeros for 'up' and 'down' counts
    df_reg = landmark_genes.copy()
    df_reg['up'] = 0
    df_reg['down'] = 0

    # Filter the gene expression data for the current drug
    filtered_up = gene_expression_df_10um_conc_up[gene_expression_df_10um_conc_up[0] == drug]
    filtered_down = gene_expression_df_10um_conc_down[gene_expression_df_10um_conc_down[0] == drug]

    # Flatten the arrays of up-regulated and down-regulated genes
    array_up = filtered_up.iloc[:, 1:].to_numpy().flatten()
    array_down = filtered_down.iloc[:, 1:].to_numpy().flatten()

    # Count the occurrences of each gene in the 'up' and 'down' arrays
    up_counts = pd.Series(array_up).value_counts()
    down_counts = pd.Series(array_down).value_counts()

    # Update the 'up' and 'down' columns in the df_reg DataFrame
    df_reg['up'] = df_reg[0].map(up_counts).fillna(0)
    df_reg['down'] = df_reg[0].map(down_counts).fillna(0)

    # Normalize the counts by the total number of occurrences of the drug
    drug_count = drug_counts.get(drug, 1)  # Default to 1 to avoid division by zero
    df_reg.iloc[:, 1:] = df_reg.iloc[:, 1:] / drug_count

    # Convert the DataFrame to a numpy array and add to the results list
    data_reg_list.append(df_reg.iloc[:, 1:].to_numpy())

# Example to verify the result
print(data_reg_list[0])
```

```python
data = np.stack(data_reg_list)
data.shape
```
