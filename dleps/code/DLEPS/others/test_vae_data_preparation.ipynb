```python
import pandas as pd
```

```python
# Load test drugs
drugs = pd.read_csv('../../data/test_drugs.csv')
drugs
```

```python
# Select the drugs from 25646 - 26000
drugs_1000 = drugs[25646:26646]
drugs_1000
```

```python
# Extract SMILES of selected drugs
import requests

drugs = list(drugs_1000['0'])
result_list = ['' for _ in range(len(drugs))]
for i in range(len(drugs)):
    # Create the URL
    url = 'https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/name/' + drugs[i] + '/property/CanonicalSMILES/TXT'

    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    try:
        if response.status_code == 200:
            # Extract the result and append it to a list 
            result_list[i] = response.text.strip()
        else:
            print('Failed to get data from URL')
    except:
        print("Unexpected error occurred")

print(len(result_list))
```

```python
smiles_df = pd.DataFrame({'drug_name' : drugs, 'SMILES' : result_list})
smiles_df
```

```python
from rdkit_transformation import standardize
standardized_SMILES = []

for smiles in smiles_df['SMILES']:
    standardized_SMILES.append(standardize(smiles))

smiles_df['standardized_SMILES'] = standardized_SMILES
smiles_df
```

```python
# Remove na
smiles_df_copy = smiles_df.copy()
smiles_df_copy_dropna = smiles_df_copy.dropna(subset=['standardized_SMILES']).reset_index(drop=True)
smiles_df_copy_dropna
```

```python
from functools import reduce

def xlength(y):
    return reduce(lambda sum, element: sum + 1, y, 0)
```

```python
def get_zinc_tokenizer(cfg):
    long_tokens = [a for a in list(cfg._lexical_index.keys()) if xlength(a) > 1] ####
    replacements = ['$','%','^'] # ,'&']
    assert xlength(long_tokens) == len(replacements) ####xzw
    for token in replacements: 
        assert token not in cfg._lexical_index ####
    
    def tokenize(smiles):
        for i, token in enumerate(long_tokens):
            try:
                smiles = smiles.replace(token, replacements[i])
            except AttributeError:
                print("Error occurred.")
        tokens = []
        try:
            for token in smiles:
                try:
                    ix = replacements.index(token)
                    tokens.append(long_tokens[ix])
                except:
                    tokens.append(token)
        except TypeError:
            print("Error occurred.")
        return tokens
    
    return tokenize
```

```python
import zinc_grammar
import nltk

_tokenize = get_zinc_tokenizer(zinc_grammar.GCFG)
_parser = nltk.ChartParser(zinc_grammar.GCFG)
_productions = zinc_grammar.GCFG.productions()
_prod_map = {}
for ix, prod in enumerate(_productions):
    _prod_map[prod] = ix
MAX_LEN = 277
_n_chars = len(_productions)
```

```python
smiles_rdkit = list(smiles_df_copy_dropna['standardized_SMILES'])
smiles_rdkit
```

```python
import numpy as np
""" Encode a list of smiles strings into the latent space """
# training_drugs_SMILES['standardized_SMILES']
assert type(smiles_rdkit) == list
tokens = map(_tokenize, smiles_rdkit)
parse_trees = []
i = 0
badi = []
for t in tokens:
    #while True:
    try:
        tp = next(_parser.parse(t))
        parse_trees.append(tp)
    except:
        print("Parse tree error at %d" % i)
        badi.append(i)
    i += 1
    #print(i)
productions_seq = [tree.productions() for tree in parse_trees]
indices = [np.array([_prod_map[prod] for prod in entry], dtype=int) for entry in productions_seq]
one_hot = np.zeros((len(indices), MAX_LEN, _n_chars), dtype=np.float32)
for i in range(len(indices)):
    num_productions = len(indices[i])
    if num_productions > MAX_LEN:
        print("Too large molecules, out of range")
    #print("i=  {%d} len(indices)=  {%d} num_productions = %d " % (i,len(indices),num_productions))
        one_hot[i][np.arange(MAX_LEN),indices[i][:MAX_LEN]] = 1.
    else:    
        one_hot[i][np.arange(num_productions),indices[i]] = 1.
        one_hot[i][np.arange(num_productions, MAX_LEN),-1] = 1.
```
