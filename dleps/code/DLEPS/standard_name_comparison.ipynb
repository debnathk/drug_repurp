# Here, we will find the standard names of the training set drug names also present in rna-seq dataset

```python
import pandas as pd
```

```python
# Discarding the training_dataset_cleaning.ipynb operations for SMILES cleaning, do the oprations here, keeping the data columns as it is
 
```

```python
# Load training dataset
df_train = pd.read_csv('../../data/training_dataset_clean.csv').drop_duplicates('DRUG NAME').drop_duplicates('Item Name').drop_duplicates('SMILES').reset_index(drop=True)
df_train
```

```python
# Check NaN values in SMILES
sum(df_train['SMILES'].isna()), sum(df_train['DRUG NAME'].isna())
```

```python
# Select only the columns - DRUG NAME, SMILES
selection = ['DRUG NAME', 'SMILES', 'standardized_SMILES']
df_selection = df_train[selection]
df_selection
```

```python
# Load the rna seq drug names
df_drug_names = pd.read_csv('../../data/gene_expression_drug_names.csv')
df_drug_names['DRUG_NAME'] = df_drug_names['DRUG_NAME'].apply(lambda x: x.lower())
df_drug_names
```

## Merge datasets by comparing drug names among the datasets

```python
df_merged_name = pd.merge(df_selection, df_drug_names, left_on='DRUG NAME', right_on='DRUG_NAME').drop_duplicates()
df_merged_name
```

## Searched till drug 1001, continuing...

```python
print(f'As of now, {len(df_merged_name)} drugs that are common in both the datasets, as per preliminary name comparison.')
```

```python
# Select unavailable drugs from training dataset
un_drug_names = sorted([i for i in df_selection['DRUG NAME'].to_list() if i not in df_merged_name['DRUG NAME'].to_list()])
print(f'Uncommon drugs: {len(un_drug_names)}')
un_drug_names
```

```python
# Filter the training dataset with common drug names
df_final = pd.merge(df_train, df_merged_name, on='DRUG NAME')
df_final
```

```python
# Save the dataset as csv
df_final.to_csv('../../data/training_dataset_clean_2.csv', index=False)
```

**Above dataset is ready to be processed for grammar vector preparation**
