```python
from selenium import webdriver
import time

def open_link_in_new_tab(driver, url):
    # Open the URL in a new tab
    driver.execute_script("window.open('{}', '_blank')".format(url))

# Create a list to store the URLs of the targets
target_urls = [f'https://www.uniprot.org/uniprotkb?query={target}' for target in unique_targets['Drug target']]

# Launch Chrome browser
driver = webdriver.Chrome()

# Open the first link in a new tab
open_link_in_new_tab(driver, target_urls[0])

# Loop through the remaining URLs
for url in target_urls[1:]:
    # Wait until the current tab is closed
    while len(driver.window_handles) > 1:
        time.sleep(1)

    # Open the next link in a new tab
    open_link_in_new_tab(driver, url)

# Wait for 2 minutes before closing the last tab
time.sleep(120)

# Close the last tab
driver.switch_to.window(driver.window_handles[-1])
driver.close()

# Close the main browser window
driver.quit()
```
